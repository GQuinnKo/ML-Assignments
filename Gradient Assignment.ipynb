```python
import numpy

w = numpy.transpose([0, -4])
step = 0.005
result = False
for i in range(10000):
    f1 = ((w[0]+step)**2 + (w[1]) - 11)**2 + ((w[1])**2 + (w[0]+step) - 7)**2
    f2 = ((w[0]-step)**2 + (w[1]) - 11)**2 + ((w[1])**2 + (w[0]-step) - 7)**2
    grad0 = (f1-f2)/(2*step)
    
    f1 = ((w[0])**2 + (w[1]+step) - 11)**2 + ((w[1]+step)**2 + (w[0]) - 7)**2
    f2 = ((w[0])**2 + (w[1]-step) - 11)**2 + ((w[1]-step)**2 + (w[0]) - 7)**2
    grad1 = (f1-f2)/(2*step)
    
    grad = numpy.transpose([grad0, grad1])
    w = w - step*grad
    
    norm = numpy.linalg.norm(grad,2)
    if norm < (10**(-12)):
        result = True
        break
if result:
    print("The algorithm converged to:", w)
    print("The algorithm took ", i+1, " iterations.")
else:
    print("The algorithm was a failure.")
    
```

```python
from scipy.stats import multivariate_normal

u = numpy.transpose([-1, 0, 4, 1, 0.5])
eNplot = numpy.empty([7,])
Nplot = numpy.empty([7,])
for j in range(1,7):
    N = 10**j
    Nplot[j-1] = N
    X = multivariate_normal.rvs(mean=u, cov=2, size = N)
    print("N = ", N)
    uN = numpy.sum(X, 0)/N
    eN = numpy.linalg.norm((uN-u), 2)**2
    eNplot[j-1] = eN
    print("Error:",eN)

f = pyplot.figure()
pyplot.loglog(Nplot,eNplot)
pyplot.xlabel("N")
pyplot.ylabel("Error")
```
